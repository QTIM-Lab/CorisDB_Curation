```python

# Read the two CSV files into Pandas DataFrames
df1 = pd.read_csv('/projects/coris_db/postgres/queries_and_stats/queries/Glaucoma/OCT_CDR_CFP_Merge/lateral_cropped_all_swin_rel_unique_filename.csv')
df2 = pd.read_csv('/projects/coris_db/postgres/queries_and_stats/queries/Glaucoma/OCT_CDR_CFP_Merge/all_raw_filename.csv')

# Specify the column on which you want to join
# Assuming 'common_column' is the column name common in both files
common_column = 'filename'

# Perform the join operation
# Assuming you want an inner join, change 'how' parameter accordingly for other types of joins
result = pd.merge(df1, df2, on=common_column, how='inner', suffixes=('_lateral', ''))

# Print the result or do further operations with it
print(result.columns)
```
