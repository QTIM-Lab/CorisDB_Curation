```python
# Sort DataFrame by filename and value_datetime
df_interp = average_df.sort_values(['filename', 'oct_datetime'])
df_interp['oct_before_date'] = None
df_interp['oct_after_date'] = None
df_interp['oct_before_value'] = None
df_interp['oct_after_value'] = None

# Group by filename
grouped = df_interp.groupby('filename')

# Function to perform linear interpolation
def interpolate(group):
    # If there's only one value, return it directly
    cfp_date = group['cfp_datetime'].iloc[0]
    if len(group) == 1:
        return group[['filename', 'cfp_datetime', 'verticalcd', 'oct_before_date', 'oct_after_date', 'oct_before_value', 'oct_after_value']].iloc[0]
    elif cfp_date in group['oct_datetime'].values:
        matching_row = group[group['oct_datetime'] == cfp_date]
        if not matching_row.empty:
            return matching_row[['filename', 'cfp_datetime', 'verticalcd', 'oct_before_date', 'oct_after_date', 'oct_before_value', 'oct_after_value']].iloc[0]
    # Find the two nearest value_datetimes
    group = group.reset_index(drop=True)
    cfp_date = group['cfp_datetime'].iloc[0]
    nearest_dates = group['oct_datetime'].iloc[[0, -1]]
    
    # Determine the difference in days between cfp_date and the two nearest dates
    days_diff = abs(nearest_dates - cfp_date).dt.days

    # print(days_diff.iloc[1])
    
    # Calculate weights for linear interpolation
    weight_total = days_diff.sum()
    weights = [(days_diff.iloc[1] / weight_total), (days_diff.iloc[0] / weight_total)]
    
    # Get corresponding values for the nearest dates
    nearest_values = group.loc[group['oct_datetime'].isin(nearest_dates), 'verticalcd']
    
    # Interpolate between the two nearest values
    interpolated_value = sum(weights[i] * nearest_values.iloc[i] for i in range(2))
    
    return pd.Series([group['filename'].iloc[0], cfp_date, interpolated_value, nearest_dates.iloc[0], nearest_dates.iloc[1], nearest_values.iloc[0], nearest_values.iloc[1]], index=['filename', 'cfp_datetime', 'verticalcd', 'oct_before_date', 'oct_after_date', 'oct_before_value', 'oct_after_value'])

# Apply interpolation function to each group
interpolated_df = grouped.apply(interpolate)

# Reset index
interpolated_df.reset_index(drop=True, inplace=True)
```
